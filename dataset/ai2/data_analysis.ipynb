```python
score_values_1_5 = {"F": 1, "T": 5, "?": 3}
score_values_0_100 = {"F": 0, "T": 100, "?": 50}
```

```python
def get_prob_score_from_labels1(labels):
    labels = labels.split()
    prob1 = []
    for i in range(len(labels)):
        prob1.append(score_values_1_5[labels[i]])
    return sum(prob1)/len(prob1)

def get_prob_score_from_labels2(labels):
    labels = labels.split()
    prob2 = []
    for i in range(len(labels)):
        prob2.append(score_values_0_100[labels[i]])
    return sum(prob2)/len(prob2)
```
